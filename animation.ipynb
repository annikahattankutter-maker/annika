```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib.animation import FuncAnimation, PillowWriter
```

```python
fig , ax = plt.subplots()
x = np.linspace(0,2 * np.pi, 100)
def sin(x):
    return np.sin(x)

line, = ax.plot(x, sin(x), lw = 2)
def update(frame):
    line.set_ydata(np.sin(x + frame / 10))
    return line,

anim = FuncAnimation(fig, update, frames = 100, interval = 50, blit = True)
writer = PillowWriter(fps = 10)
anim.save('sine_wave.gif', writer = writer)
```

```python
fig1, ax1 = plt.subplots()
ax1.set(xlim = [0, 5], ylim = [-5, 10])
t = np.linspace(0, 10, 100)

# variables
g = -9.81
v0 = 10


height = v0 * t + (1/2) * g * t**2
line = ax1.plot(t[0], height[0])[0]

def update(frame):
    line.set_xdata(t[:frame])
    line.set_ydata(height[:frame])
    return line

ani = FuncAnimation(fig1, update, frames = 40, interval = 30)
writer = PillowWriter(fps = 10)
ani.save('Projectile.gif', writer = writer)
```
