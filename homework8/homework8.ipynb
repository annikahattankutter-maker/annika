```python
lat_med = df['latitude'].median()
long_med = df['longitude'].median()
print(lat_med)
print(long_med)
```

```python
abs_value_lat = (df['latitude'] - lat_med).abs()
df['distance from med latitude'] = abs_value_lat

abs_value_long = (df['longitude'] - long_med).abs()
df['distance from med longitude'] = abs_value_long
df
```

```python
print(df.shape)
```

```python
closest_lat = df.loc[df['distance from med latitude'].idxmin()]
farthest_lat = df.loc[df['distance from med latitude'].idxmax()]

closest_long = df.loc[df['distance from med longitude'].idxmin()]
farthest_long = df.loc[df['distance from med longitude'].idxmax()]

print(f"{closest_lat['state']} is closest state to median latitude distance by {closest_lat['distance from med latitude']:.2f}")
print(f"{farthest_lat['state']} is farthest state away from median latitude distance by {farthest_lat['distance from med latitude']:.2f}")

print(f"{closest_long['state']} is closest state to median longitude distance by {closest_long['distance from med longitude']:.2f}")
print(f"{farthest_long['state']} is farthest state away from median longitude distance by {farthest_long['distance from med longitude']:.2f}")
```

```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = sns.load_dataset('planets')
df = df.dropna()
df
```

```python
df.shape # says it's (498, 6) but i think it's actually (785, 6)
# 498 rows, 6 columns
df.head(5)
```

```python
df.tail(5)
# method = method used to discover the planet
# number = number of planets
# orbital_period = orbital period of planet (doesn't state the units)
# mass = mass of planet
# distance = distance from earth? probably in light years
# year = year it was discovered?
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
fig, axes = plt.subplots(2,2, figsize = (16,16))


axes[0, 0].scatter(df['orbital_period'], df['mass'], color = 'red', marker = '+', label = 'data')
axes[0, 0].set_xlabel('Orbital Period (Days)')
axes[0, 0].set_ylabel('Mass (Jupiter Masses)')
axes[0, 0].set_title("Planet's Orbital Period vs Its Mass", fontsize = 20)
axes[0, 0].legend()
# learned that the relationship between mass and orbital period is directly proportional

axes[0, 1].scatter(df['distance'], df['year'], color = 'green', marker = '+')
axes[0, 1].set_xlabel('Distance (Light Years)')
axes[0, 1].set_ylabel('Year of Discovery')
axes[0, 1].set_title("Planet's Distance vs Year of Discovery", fontsize = 20)
# learned planets farther away were discovered more recently

sns.histplot(df['mass'], ax = axes[1, 0], color = 'blue')
axes[1, 0].set_xlabel('Mass (Jupiter Masses)')
axes[1, 0].set_ylabel('Count')
axes[1, 0].set_title("Number of Planets with a Certain Mass", fontsize = 20)
# learned that most planets have low mass than high mass

sns.boxplot(df['distance'], ax = axes[1, 1], color = 'purple')
axes[1, 1].set_ylabel('Distance (Light Years)')
axes[1, 1].set_title("Box Plot of Distance Data", fontsize = 20)
# learned that most exoplanets in the data are around 50 light years away


plt.show()
```
