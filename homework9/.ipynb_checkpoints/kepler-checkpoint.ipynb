```python
row_orbital_period = df.loc[10][1:].astype(float)
row_orbital_period
```

```python
row_perihelion = df.loc[8][1:].astype(float)
row_perihelion
```

```python
row_aphelion = df.loc[9][1:].astype(float)
row_aphelion
```

```python
# semi major axis = half of the major axis, or the largest radius of an eclipse
# Perihelion = point in a planet's orbit that is closest to the sun
# Aphelion = point in a planet's orbit that is farthest from the sun
```

```python
semi_major_axis = (row_perihelion + row_aphelion) / 2
semi_major_axis
```

```python
semi_major_axis_row = pd.Series(
    ['Semi Major Axis (10^6 km)'] + list(semi_major_axis),
    index = df.columns)

df = pd.concat([df, semi_major_axis_row.to_frame().T], ignore_index = True)
df
```

```python
df.shape
```

```python
y = row_orbital_period
x = semi_major_axis



plt.figure()
plt.scatter(x, y, label = 'data', marker = '+', c = 'red')
plt.xlabel('Semi Major Axis (10^6 km)')
plt.ylabel('Orbital Period (days)')
plt.title("Planet's Semi Major Axis vs. Orbital Period")
plt.legend()
plt.show()
```

```python
def keplers_third_law(x, m):
    return m * x ** 1.5
    
p0 = [1]
popt, pcov = curve_fit(keplers_third_law, x, y, p0=p0)
m_fit = popt

residuals = y - keplers_third_law(x, m_fit)
chi2 = np.sum(residuals**2)
dof = len(x) - len(popt)
chi2_reduced = chi2 / dof
print(f"Reduced Chi-squared: {chi2_reduced:.2f}")

plt.figure(figsize = (8, 6))
plt.scatter(x, y, label='Planets', color='blue')
plt.plot(x, keplers_third_law(x, m_fit), color='red', label='Best Fit')

for planet in semi_major_axis.index:
    plt.text[semi_major_axis[planet]]
plt.title('Semi Major Axis vs. Orbital Period for Planets, Reduced Chi-squared = 41276.71')
plt.xlabel('Semi Major Axis (10^6 km)')
plt.ylabel('Orbital Period (days)')

plt.grid()
plt.legend()
plt.show()
```
