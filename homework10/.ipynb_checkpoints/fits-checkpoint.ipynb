```python
import matplotlib.pyplot as plt
from astropy.io import fits
import numpy as np
```

```python
file_list = ['M13_0001.fits', 'M13_0002.fits', 'M13_0003.fits', 'M13_0004.fits', 'M13_0005.fits'] 

all_headers = []
all_data = []

for file in file_list:
    with fits.open(file) as hdul:
        header = hdul[0].header
        data = hdul[0].data
        all_headers.append(header)
        all_data.append(data)
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    axes[i].imshow(all_data[i], cmap='inferno', origin='lower')
    axes[i].set_title(f"Image {i+1}")
    axes[i].axis('off')

plt.tight_layout()
plt.show()
```

```python
vmin = min(np.min(img) for img in all_data)
vmax = max(np.max(img) for img in all_data)

fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    im = axes[i].imshow(all_data[i], cmap='inferno',
                        origin='lower',
                        vmin=vmin, vmax=vmax)

    axes[i].set_title(f"Image {i+1}")
    axes[i].axis('off')
    fig.colorbar(im, ax=axes[i], shrink=0.7)

plt.tight_layout()
plt.show()
# notice that all the graphs look almost identical and are all the same color
```

```python
fig, axes = plt.subplots(1, 5, figsize=(20, 4))

for i in range(5):
    axes[i].hist(all_data[i].ravel(), bins=50, color='gray')
    axes[i].set_title(f"Histogram {i+1}")
    axes[i].set_xlabel("Pixel Value")
    axes[i].set_ylabel("Count")

plt.tight_layout()
plt.show()
```

```python
for i, img in enumerate(all_data, 1):
    mean_val = np.mean(img)
    median_val = np.median(img)
    print(f"Image {i}: mean = {mean_val:.2f}, median = {median_val:.2f}")
# a. x-axis = pixel value, y-axis = how many pixels there are
# b. bins = 50 means that the x-axis data is divided into 50 bars
# c. most is centered aroun 500 pixels
# d. data is pretty much all centered around the 500 pixel mark
```

```python
stacked_images = np.array(all_data)
combined_image = np.median(stacked_images, axis=0)

combined_median = np.median(combined_image)
combined_mean = np.mean(combined_image)

plt.figure(figsize=(6,4))
plt.hist(combined_image.ravel(), bins=50, color='gray')
plt.xlabel("Pixel Value")
plt.ylabel("Count")
plt.title("Histogram of Combined Image")
plt.show()
```

```python
pixels = combined_image.ravel()
counts, bin_edges = np.histogram(pixels, bins=50)
vmin = np.median(pixels) - 50
vmax = np.median(pixels) + 50

plt.figure(figsize=(6,6))
plt.imshow(combined_image, cmap='inferno', origin='lower', vmin=vmin, vmax=vmax)
plt.colorbar(label="Pixel Value")
plt.title("Combined Image")
plt.show()
```

```python
# the image is a lot more detailed than the other ones.
fits.writeto("combined.fits", combined_image, overwrite=True)
```
