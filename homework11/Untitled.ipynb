```python
df = df[~df.apply(lambda row: row.astype(str).str.contains('\n').any(), axis=1)]
```

```python
from astropy.table import Table

data_table = Table.read('global_CCl4_MM.dat', format='ascii')
```

```python
data_table
```

```python
columns_to_keep = ['CCl4ottoyr', 'CCl4ottoGLm', 'CCl4ottoGLsd']

df = data_table[columns_to_keep].to_pandas()
```

```python
x = df['CCl4ottoyr']                                 
y = df['CCl4ottoGLm']            
yerr = df['CCl4ottoGLsd']      

plt.figure(figsize=(10,6))
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='red', capsize=3, markersize=4)
plt.xlabel('Date')
plt.ylabel('Global Mean Concentration')
plt.title('Global Mean Concentration with Error Bars')
plt.grid(True)
plt.show()
```

```python
from scipy.optimize import curve_fit

x = df['CCl4ottoyr'].values                                 
y = df['CCl4ottoGLm'].values            
yerr = df['CCl4ottoGLsd'].values      


mask = ~np.isnan(x) & ~np.isnan(y) & ~np.isnan(yerr)
x = x[mask]
y = y[mask]
yerr = yerr[mask]

x_centered = x - np.mean(x)

def linear_model(x, m, b):
    return m*x + b
popt, pcov = curve_fit(linear_model, x_centered, y, sigma=yerr)
m, b = popt
y_fit = linear_model(x_centered, m, b)

plt.figure(figsize=(10,6))
plt.errorbar(x, y, yerr=yerr, fmt='o', ecolor='red', capsize=3, markersize=4, label='Data')
plt.plot(x, y_fit, 'k-', label=f'Fit: y = {m:.2e}x + {b:.2f}')
plt.xlabel('Date')
plt.ylabel('Global Mean Concentration')
plt.title('Negative Linear Fit using curve_fit')
plt.legend()
plt.grid(True)
plt.show()
```

```python
N = len(y)
chi2_red = np.sum(((y_fit - y) / yerr)**2) / (N - 1)
print(f"Reduced chi-square: {chi2_red:.2f}")
```

```python
# Since the reduced chi-squared is 13.14, it is considered not a good fit.
# In my opinion, the line of best fit looks pretty accurate though.
```
