```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
data = np.loadtxt("../../course_assignments/homework7/star_data.txt", skiprows=1)
data
```

```python
np.shape(data) # 2642 rows
```

```python
magnitude = data[:, 0]
temperature = data[:, 1]

print(magnitude.max()) # dimmest star
print(magnitude.min()) # brightest star

print(temperature.max()) # hottest star
print(temperature.min()) # coolest star
```

```python
log_magnitude = np.log(magnitude)
log_temperature = np.log(temperature)

plt.plot(figsize = (8, 6))
plt.scatter(log_temperature, log_magnitude)

plt.gca().invert_xaxis()
plt.gca().invert_yaxis()

plt.show()
```
