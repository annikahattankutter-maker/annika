```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 2*np.pi, 1000)
def make_sine_wave(z, A, w):
    y = A*np.sin(z*w)
    return y

y = make_sine_wave(x, 1, 1)
plt.plot(x, y)
plt.show()
```

```python
amp = np.array([0.5, 1, 1.5, 2, 2.5])
freq = np.array([1, 2, 3, 4, 5])
amp
```

```python
freq
```

```python
fig, axs = plt.subplots(1,3, figsize = (25, 5))

for amplitude, frequency in zip(amp, freq):
    y1 = make_sine_wave(x, amp, 1)
    axs[0].plot(x, y1)
    y2 = make_sine_wave(x, 1, freq)
    axs[1].plot(x, y2)
    y3 = make_sine_wave(x, amp, freq)
    axs[2].plot(x, y3)
    
plt.show()
```
